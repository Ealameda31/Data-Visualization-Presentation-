# Scatterplots

```python
#100 random numbers with in sets of 3

Scatter_Example= pd.read_excel("Stuff.xlsx")

Array=np.array(np.random.random((100,3)))
plt.scatter(Array[:,0], Array[:,1])

plt.title("Observations")
plt.xlabel("East")
plt.ylabel("North")

x=plt.scatter(Array[:,0],Array[:,1], c=Array[:,2])
cbar=plt.colorbar()
cbar.set_label("elevation (m)", labelpad=20)
x.set_clim(0,1)


plt.show()
```

```python
colors = ['g','r']

Test1 = plt.scatter(Scatter_Example["Test 1"][0], Scatter_Example["Test 1"][1], marker='x', color='red', s=200)
Test2 = plt.scatter(Scatter_Example["Test 2"][0], Scatter_Example["Test 2"][1], marker='o', color="green", s=200)
Test3 = plt.scatter(Scatter_Example["Test 3"][0], Scatter_Example["Test 3"][1], marker='+', color="orange",s=200)
Test4 = plt.scatter(Scatter_Example["Test 4"][0], Scatter_Example["Test 4"][1], marker='*', color="blue", s=200)
Test5 = plt.scatter(Scatter_Example["Test 5"][0], Scatter_Example["Test 5"][1], marker='s', color="purple", s=200)


plt.legend((Test1, Test2,Test3,Test4,Test5),
           ('Test1', 'Test2','Test3','Test4','Test5'),
           scatterpoints=1,
           loc=9,
           ncol=2,
           fontsize=12, bbox_to_anchor=(1.3,0.7))

plt.title("Test Comparisons")

plt.xlabel("Mean")

plt.ylabel("Variance")


#plt.grid()
plt.show()
```
